### Importing Library and Datasets

```python
#as this is a big dataset we can drop Null Values 
data.dropna(inplace=True)
data.isnull().sum()
```

```python
#Now lets check the frequency of products
product_count = data.groupby(['Product_Code']).size().reset_index(name='counts').sort_values(['counts'],ascending=False)
product_count.head(10)
```

##### Since we are doing time forecasting hence we can take any one of the product and do time forecasting on it. As Product_1359 has highest count let's do Time Forecasting on it.

```python
#Create a new data frame with only product_code 1359
product_1359 = data.loc[data['Product_Code'] == 'Product_1359']

product_1359.head()
```

```python
#Find details of the new data frame
product_1359.describe()
```

##### As we can see Product_Code 1359  has only category of product(Category_019) and 16936 Order_Demand. Let's begin our Time Forecasting now on this.

```python
#Delete redundats columns for time forecasting
product_1359 = product_1359.drop(columns=["Warehouse", "Product_Code", "Product_Category"])

product_1359.head()
```

### Feature Engineering Begins

```python
#Convert the "Date" to date format
product_1359.loc[:, 'Date'] = pd.to_datetime(product_1359.loc[:,'Date'], format="%Y/%m/%d")
product_1359.describe()
```

```python
#Indexing with timeseries data
product_1359 = product_1359.set_index('Date')
product_1359.index
```

```python
product_1359['Order_Demand'] = product_1359['Order_Demand'].astype(str)
product_1359['Order_Demand'] = product_1359['Order_Demand'].map(lambda x: x.lstrip('(').rstrip(')'))
product_1359['Order_Demand'] = product_1359['Order_Demand'].astype(int)
```

```python
#Calculating sum of order demand on per month basis
product_1359_Month = product_1359["Order_Demand"].resample('MS').sum()
product_1359_Month['2016':]
```

```python
#Ignoring date 2017-01-01 as it is incomplete for the year
product_1359_Month = product_1359_Month.drop(pd.to_datetime('2017-01-01'))
product_1359_Month['2017':]
```

```python
#Visualizing Demand Time Series Data
product_1359_Month.plot(figsize=(15, 6))
plt.show()
```

### Time Series Forecasting begins

#### Using FbProphet Model

```python
#reference : https://facebook.github.io/prophet/docs/quick_start.html#python-api 
product_1359 = pd.DataFrame({'Date':product_1359_Month.index, 'Order_Demand':product_1359_Month.values})
product_1359.columns = ['ds', 'y']
product_1359.head()
```

```python
#fitting the model
model = Prophet(interval_width=0.95, weekly_seasonality=True)
model.fit(product_1359)
```

```python
#making future model for 365 days ie dor 2017-18
future = model.make_future_dataframe(periods=365)
future.tail()
```

```python
#predicting the future model
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plotting the demand graph on basis of the model
plt.figure(figsize=(18, 6))
model.plot(forecast, xlabel = 'Date', ylabel = 'Demand')
plt.title('Product 1359 Demand');
```

```python
#componenets analysis
fig2 = model.plot_components(forecast)
```

### Conclusion: 
After using the fbprophet time series forecasting model for Product_1359 following observations can be made:

Demand for this product is increasing from 2012 to 2018

As for the yearly trend, there is a maximum increase in demand from July to September

For the weekly trend, demand is highest on a Tuesday and then goes to negative on a Friday
